```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import math
from google.colab import drive
#drive.mount('/content/drive')
```

```python
# display ori image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Image Processing Files Data/Lab2/tree.jfif',cv2.IMREAD_GRAYSCALE)
cv2_imshow(image)
```

```python


x = np.size(image)

print(x)
#plt.hist(image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k') #calculating histogram
[counts, pixel] = np.histogram(image, 256)

plt.subplot(121)
plt.imshow(image, cmap = 'gray')
plt.title('Original Image')
plt.subplot(122)
plt.bar(pixel[:-1], counts, color = 'red')
plt.title('Histogram of image')
plt.show()





```

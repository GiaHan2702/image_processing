Contrast/ Brightness edit image

```python
# EX2. Brightness and contrast enhancement
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# load test image
image = cv2.imread('sea_grayscale.jpg')
# convert to float and normalize it
image = image.astype(np.float32)/255.0
```

```python
# brightness enhancement
scale = 1.0 # unchange
#scale = 0.3 # darker
#scale = 1.3 # brighter
scale_image = scale*image
plt.subplot(121)
plt.imshow(image, cmap = 'gray')
plt.title('Original Image')
plt.subplot(122)
plt.imshow(scale_image, cmap = 'gray')
plt.title('Scaled Brightness Image with factor 1_ unchanged')
```

```python
# save image
cv2.imwrite('Brightness_scale_image_unchanged.jpg', scale_image)
```

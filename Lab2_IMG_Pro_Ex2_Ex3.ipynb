```python
# Image subtraction

# load img
liveImg = cv2.imread('live.jpg', cv2.IMREAD_GRAYSCALE)/255
maskImg = cv2.imread('mask.jpg',cv2.IMREAD_GRAYSCALE)/255

plt.subplot(1,2,1)
plt.imshow(liveImg, cmap ='gray')
plt.subplot(1,2,2)
plt.imshow(maskImg, cmap ='gray')

```

# New Section

```python
#calculate the difference
diffImg = np.abs(liveImg - maskImg)
gamma = 0.5

diffImg_contrasted = np.power(diffImg,gamma)
plt.subplot(121)
plt.imshow(diffImg*255, cmap ='gray')
plt.title('Difference Image')
plt.subplot(122)
plt.imshow(diffImg_contrasted*255, cmap ='gray')
plt.title('Different Image contrasted')
```

```python
# Ex1. Quantization
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# display color image
color_image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Image Processing Files Data/nature.jpg',cv2.IMREAD_COLOR)
cv2_imshow(color_image);
cv2.imwrite('nature_color.png', color_image);
```

```python
# display grayscale image
grayscale_Img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Image Processing Files Data/nature.jpg',cv2.IMREAD_GRAYSCALE);
cv2_imshow(grayscale_Img)
```

```python
# c) Quantize grayscale image in 1-8 bits
# convert image to floating points format and normalize it

image = grayscale_Img.astype(np.float32)/255.0 # change to float and / max value is 255 => dua ve thang 0 -> 1

# repeat the quantization with different bits
for NumBits in range(1,9):
  # quantize to a given number of bits
  num_level = 2** NumBits
  level_gap = 256/ num_level

  # quantization
  quantized_img = np.uint8(np.ceil(grayscale_Img/level_gap)*level_gap-1) # moi buoc =1  nac nhan gia tri cua no

  # plot image
  plt.subplot(2,4,NumBits)
  plt.imshow(quantized_img, cmap = "gray")
  # image caption and save
  if NumBits == 1:
    name = str(NumBits) + 'bit'
  else:
    name = str(NumBits) + 'bits'
  plt.title(name)
  plt.savefig('Quantization_in_'+ name + 'j.jpg')
plt.show()



```
